```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('ques_1.csv')
```

```python
df
```

```python
grouped_df = df.groupby('country')['sales_amount'].sum().reset_index()
```

```python
sorted_df = grouped_df.sort_values('sales_amount', ascending=False)
```

```python
plt.figure(figsize=(20,5)) # Adjust the figure size as per your preference
plt.bar(sorted_df['country'], sorted_df['sales_amount'])
plt.xlabel('Total Revenue')
plt.ylabel('Country')
plt.title('Revenue from Different Countries in the DVD Store')
plt.yicks(rotation=90)
plt.gca().xaxis.set_tick_params(pad=25)  # Increase the separation between xticks
plt.show()
```
